## Add CVE Description to CISA KEV
"CVE","vulnerabilityName"
vulnerabilityName gives the summary impact / weakness

```python
df_cisakev = pd.read_csv('data_in/known_exploited_vulnerabilities.csv', usecols=["cveID", "vulnerabilityName"])
df_cisakev
```

```python
df_cisakev.rename(columns={'cveID': 'CVE'}, inplace=True)
df_cisakev
```

```python
df_cisakev = df_cisakev.merge(df_cve[['CVE', 'Description']], on='CVE', how='left')
df_cisakev
```

```python
# should be zero without Description
df_cisakev.Description.isna().sum() 
```

```python
df_cisakev.to_csv("./data_out/cisakev_with_cve_descriptions.csv", index=False, quoting=csv.QUOTE_ALL, escapechar='\\')
```
